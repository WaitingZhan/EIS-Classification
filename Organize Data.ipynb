```python
print(files)
```

```python
freqency_matrix=[]
real_matrix=[]
imaginary_matrix=[]
```

```python
number_file = 0
for file in files:
    number_file = number_file + 1
    df = pd.read_csv(file,skiprows = 18,sep = '\s+')
    freqency=df.iloc[:,1]
    real=df.iloc[:,2]
    imaginary=df.iloc[:,3]
    freqency_matrix.append(freqency)
    real_matrix.append(real)
    imaginary_matrix.append(imaginary)
    #print(number_file)
```

```python
rm = pd.DataFrame(real_matrix)
```
