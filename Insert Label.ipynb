#  Label the data  
Read the data from category fold, label them with  LSC as 1, LSCF as 2, SBSC as 3
![title](.\Figure\)

```python
pathLSC ='./NewData/category/LSC.csv'
```

```python
inssertlable = pd.read_csv(pathLSC)
```

```python
inssertlable.insert(0, "label", "1")
```

```python
inssertlable.head()
```

```python
LSC = inssertlable.drop(inssertlable.columns[1:4], axis=1)
```

```python
LSC.to_csv("./NewData/label/LSClabel1.csv")
print("Data save to ./NewData/label/LSClabel1.csv")
```

## 2 LSCF

```python
pathLSCF ='./NewData/category/LSCF.csv'
```

```python
inssertlable2 = pd.read_csv(pathLSCF)
inssertlable2.insert(0, "label", "2")
inssertlable2.head()
```

```python
LSCF = inssertlable2.drop(inssertlable2.columns[1:4], axis=1)
```

```python
LSCF.head()
```

```python
LSCF.to_csv("./NewData/label/LSCFlabel2.csv")
print("Data save to ./NewData/label/LSCFlabel2.csv")
```

## 3.  SBSC

```python
pathSBSC ='./NewData/category/SBSC.csv'
inssertlable3 = pd.read_csv(pathSBSC)
inssertlable3.insert(0, "label", "3")
inssertlable3.head()
```

```python
SBSC = inssertlable3.drop(inssertlable3.columns[1:4], axis=1)
```

```python
SBSC.head()
```

```python
SBSC.to_csv("./NewData/label/SBSClabel3.csv")
print("Data save to ./NewData/label/SBSClabel3.csv")
```

## 4. combine data 1 + data 2 + data 3

```python
 result = LSC.append(LSCF,sort=True).append(SBSC,sort=True)
```

```python
result.to_csv("./NewData/label/data.csv")
print("Data save to ./NewData/label/data.csv")
```
